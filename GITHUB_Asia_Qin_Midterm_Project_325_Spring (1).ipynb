
#Limitations
There may be potential bias when it comes to if regions have any underreported data or if specific groups are underrepresented in the dataset such as race, age or gender. And especially during the pandemic, there may be an increase in mail-in registration ballots in comparison to in-person registration ballots. The mail-in registrations might not have completely processed due to external factors (lost during commute),so those with the intention of voting might not have been ale to during this time. The article even mentions this: "In 2016 and in the pre-pandemic months of 2020, in-person registration at places like departments of motor vehicles made up a large plurality, or even a majority, of new registrants in the four places for which we have data on how new voters are registering. But after the pandemic caused most states to shutter many government offices, those registrations dwindled."  Moreover, some other external factors ranging from voter signature authenticity, changes in election laws, significant political events, or demographic shifts are not taken into consideration in this dataset--it's just the number of registered voters from 2016-2020. Finally, the dataset only takes into consideration the registered voters in 2016 and 2020 and not the years between it which could limit the potential trends we might pick up from in between and only provides the first half of the year.


# Exploratory Data Analysis

```python
highest_amount_voters = voters[["NAME", "Year", "Month", "registered_voters"]].sort_values(by="registered_voters",ascending = False)
highest_amount_voters
```

```python
average_voter_registration = voters['registered_voters'].mean()
average_voter_registration
# We are finding the mean of all the registered voters for each state and accounting for each year and month
```

```python
high_voters = voters.loc[voters.registered_voters > average_voter_registration].sort_values(by="registered_voters",ascending = False)
high_voters
```

Voter registration was higher than the average 36 times.

 # Each state going above the national average for new voter registration:

```python
state_above_average_count = voters[voters['registered_voters'] > voters['registered_voters'].mean()]['NAME'].value_counts()
state_above_average_count
# Texas: 8
# Florida: 7
# California: 7
# North Carolina: 6
# Illinois: 4
# Arizona: 3
# Georgia: 1
```

Three states had the highest average voter registration overall: Texas, California, and Florida

```python
highest_voter_mean=voters.groupby("NAME")["registered_voters"].mean()
highest_voter_mean.sort_values(ascending=False).head(3)

```

# Data Visualization

```python
state_voter_sum = voters.groupby('NAME')['registered_voters'].sum().reset_index()
state_voter_sum = state_voter_sum.sort_values(by='registered_voters', ascending=False)

plt.figure(figsize=(10, 10))
sns.barplot(data=state_voter_sum,x='registered_voters', y='NAME', palette="rocket")
plt.title("Total Voter Registration per State")
plt.xlabel("Total Registered Voters")
plt.ylabel("State")
plt.show()

```

The highest number of registered voters from 2016-2020 is California while the lowest is Delware. It appears that the higher population the state, the more new registered voters there will be.

```python
voters_two_year = voters[voters['Year'].isin([2016, 2020])]
plt.figure(figsize=(10,5))

colors = ["#8EC3E6","#FFA53F"]
sns.barplot(x='Month', y='registered_voters', data=voters_two_year, hue='Year', palette="crest")

plt.title("Voter Registration Comparison in 2016 vs 2020")
plt.xlabel("Month")
plt.ylabel("New Registered Voters")
plt.legend(title='Month')
plt.show()

```

 It seems that the trend of increase and decreasing rates of newly registered voters is very similar to each other. However, the year 2020 has a more drastic shift in decrease from March to April compared to in 2016.
The new voter registration was at its greatest in February in 2020 while new voter registration was at its greatest in March in 2016. The sudden change in new voters in 2020 most likely indicated the impact of COVID-19
as the pandemic worsens and quarantine begins during that time and the aforementioned limitations will start impacting the rate in which people register to vote.

# Answering 2nd Research Question:
Because California is one of the most populated states in America, how has this state's registered voting rate changed over the past year in 2016 and 2020: during the rise of COVID-19?

```python
california_voters = voters[(voters['NAME'] == 'California') & (voters['Year'].isin([2016, 2020]))]
california_voters
```

```python
colors = ["#720B98","#FFBF01"]
sns.lineplot(x='Month', y='registered_voters', hue='Year', data=california_voters, marker='o', palette=colors)
plt.title("Voter Registration Comparison in 2016 vs 2020 in California")
plt.xlabel("Month")
plt.ylabel("Number of New Registered Voters")
plt.legend(title='Year')

plt.show()

# This line plot emphasizes the significant drop from March to April specifically in the state of Califronia and compares the rate of new registration votes are with the
# rest of the month for both the years of 2016 and 2020.
```

```python
voter_per_state = voters.groupby("NAME")["registered_voters"].mean()
voter_per_state
```

```python
voting_merge = bbox.merge(voter_per_state, on="NAME", how="inner")
voting_merge
```

```python
fig=plt.figure(1,figsize=(25,15))
ax=fig.add_subplot()
voting_merge.plot(column="registered_voters",ax=ax, legend=True,cmap="Purples",linewidth=.25,edgecolor="black",
            legend_kwds={"label": "Average Voter Registration in the Year 2016 and 2020",
            "orientation": "horizontal"})
voting_merge.apply(lambda x: ax.annotate(text=x.STUSPS,
    xy=(x.geometry.centroid.x, x.geometry.centroid.y),ha="center",fontsize=12),axis=1)
plt.show()
```
